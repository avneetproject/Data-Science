MATPLOTLIB

```python
import matplotlib.pyplot as plt
# %matplotlib inline
import numpy as np
```

```python
#examples
x = np.arange(0,10)
y = np.arange(11,21)
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
#scatter plot
plt.scatter(x,y,c='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2d')
plt.savefig('Test.png')
```

```python
x = y*y
plt.plot(x,y,'go--')
```

```python
#subplot
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'bo')
```

```python
#sine curve
x = np.arange(0,4*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('sine wave form')

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('cos wave form')

plt.show()
```

```python
#bar graph
x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y) 
plt.bar(x2, y2, color = 'r') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()
```

```python
#HISTOGRAM
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a,bins=20) 
plt.title("histogram") 
plt.show()
```

```python
#BOX PLOT
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=False,patch_artist=False);  
```

```python
#PIE CHART
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0.05, 0.025, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
