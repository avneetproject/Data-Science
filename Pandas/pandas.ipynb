DAY 2 - CSV 

```python
from io import StringIO,BytesIO
data = ('col1,col2,col3\n'
        'x,y,1\n'
        'a,b,2\n'
        'c,d,3')
type(data)
pd.read_csv(StringIO(data))
```

```python
pd.read_csv(StringIO(data),index_col=False)
pd.read_csv(StringIO(data),escapechar='\n')   #this is to escape characters in between file
pd.read_csv(StringIO(data),sep=',')   #this is to to use the seperator
```
